```python
import matplotlib.pyplot as plt
import numpy as np
import math
import random


#wierzchołki trapezu
x = [-1,0,1,2,3]
y = [0,1/3,1/3,1/3,0]


def losuj(n):
    for i in range (1,n):
        x = random.random()
        #odwrotności dystrybuanty z wykładu
        if x < 1/6:
            x = math.sqrt(6*x) - 1
        elif x>=1/6 and x<5/6:
            x=3*x-1/2
        else:
            x=3-math.sqrt(6-6*x)
        tablica.append(x)


tablica = []
N = 1000
losuj(N)

#plt.plot(x,y)


plt.figure(dpi=150, clear=True)
plt.plot(x,y)
plt.hist(tablica, density=True, bins = 70)
plt.title('Dla n = 1 000')
plt.show()


N = 100000
losuj(N)



plt.figure(dpi=150, clear=True)
plt.plot(x,y)
plt.hist(tablica, density=True, bins = 70)
plt.title('Dla n = 100 000')
plt.show()
```
