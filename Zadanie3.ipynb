```python
import matplotlib.pyplot as plt
import numpy as np
import math
import random

#zakładamy lambda = 4
tablica = []
k = 16 #ilosc punktow
lamb = 4

def rozklad_oczekiwany(k):
    for i in range (0,k):
        tablica.append(poisson_wzor(i))

def poisson_wzor(i):
    result = (math.pow(lamb,i)*math.exp(-lamb))/math.factorial(i)
    return result

def generator_liczb():
    q = math.exp(-lamb)
    s = q
    p = q
    u = random.random()
    k = 0
    if u > s:
        while u>s:
            k = k + 1
            p = (p * lamb) / k
            s = s + p
        return k
    else:
        return k

    
rozklad_oczekiwany(k) #wyliczany ze wzoru

#wykres wyliczony ze wzoru
plt.plot(tablica, 'ro')
plt.grid()
plt.axis([0,14,0,0.2])
plt.title('Rozkład oczekiwany');
plt.show()    
    
    
values = []
N = 100000
for i in range (0,N):
    values.append(generator_liczb())
plt.hist(values,density = True, bins=16)
#plt.axis([0,14,0,0.2])
plt.plot(tablica, 'ro')
plt.title('Wykres wyliczony "eksperymentalnie"');
plt.show()
```
